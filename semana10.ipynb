# **Escritura de archivos desde Python**

```python
# Crear y escribir una rchivo de texto
with open("archivo.txt", "w") as f:
  f.write("Hola, este es un archivo de texto. \n")
  f.write("Segunda linea. \n")
```

Con el argumento "w" se crea el archivo si no existe. También sobreescribe el archivo si ya existe, borrando todo su contenido previo.

Con archivos.csv

```python
import csv

datos = [
    ["Nombre", "Edad", "Ciudad"],
    ["Ana", 25, "Bogotá"],
    ["Luis", 30, "Medellín"],
    ["Carlos", 28, "Cali"]
]

with open("datos.csv", "w", newline="") as f:
  writer = csv.writer(f)         #La función writer es una función integrada?
  writer.writerows(datos)
```

```python
import csv

ruta = "/content/drive/MyDrive/datos.csv"

# Leer
with open(ruta, "r") as f:
    data = list(csv.reader(f))

# Mostrar antes
print("ANTES:")
print(data)

# Agregar fila
data.append(["Diego", "28", "Armenia"])

# Guardar
with open(ruta, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)

# Mostrar después
print("\nDESPUÉS:")
with open(ruta, "r") as f:
    print(f.read())

```
